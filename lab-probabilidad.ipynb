**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
import math
from scipy.stats import norm

# Datos del problema
asientos = 450
tickets = 460
p_no_show = 0.03
p_show = 1 - p_no_show  # 0.97

# 1. Cálculo de media y desviación típica de la binomial
media = tickets * p_show
sigma = math.sqrt(tickets * p_show * p_no_show)

# 2. Corrección por continuidad: P(X ≤ 450) ≈ P(X ≤ 450.5)
x = asientos + 0.5

# 3. Cálculo del valor Z
z = (x - media) / sigma

# 4. Probabilidad acumulada en la normal estándar
probabilidad = norm.cdf(z)

# Resultado
print(f"Media (μ): {media:.2f}")
print(f"Desviación típica (σ): {sigma:.2f}")
print(f"Z score: {z:.2f}")
print(f"Probabilidad de que haya asiento para todos: {probabilidad:.4f} ({probabilidad*100:.2f}%)")
```

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python
# Datos
p_exito = 0.3
p_fallo = 1 - p_exito  # 0.7

# Probabilidad de necesitar 3 o más intentos antes de éxito
prob = p_fallo ** 2  # fallo 1 * fallo 2

print(f"Probabilidad de necesitar al menos 3 intentos: {prob:.4f} ({prob*100:.2f}%)")
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

```python
# Datos
lambda_visitas = 500
capacidad = 550

# Aproximación normal
media = lambda_visitas
sigma = math.sqrt(lambda_visitas)

# Corrección por continuidad P(X > 550) ≈ P(X > 550.5)
x = capacidad + 0.5

# Cálculo del Z-score
z = (x - media) / sigma

# Probabilidad de ser superado
prob_over = 1 - norm.cdf(z)

print(f"Media: {media}")
print(f"Sigma: {sigma:.2f}")
print(f"Z-score: {z:.2f}")
print(f"Probabilidad de servidor abrumado: {prob_over:.4f} ({prob_over*100:.2f}%)")
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
lambda_visitas = 500
capacidad = 550
horas = 24

# Aproximación normal para 1 hora
media = lambda_visitas
sigma = math.sqrt(lambda_visitas)

# Corrección por continuidad P(X > 550) ≈ P(X > 550.5)
x = capacidad + 0.5
z = (x - media) / sigma

# Probabilidad de servidor abrumado en 1 hora
p_over_1h = 1 - norm.cdf(z)

# Probabilidad de NO ser abrumado en 1 hora
p_no_over_1h = 1 - p_over_1h

# Probabilidad de NO ser abrumado en un día completo
p_no_over_day = p_no_over_1h ** horas

# Probabilidad de ser abrumado al menos una vez en el día
p_over_day = 1 - p_no_over_day

print(f"Probabilidad de servidor abrumado en 1 hora: {p_over_1h:.4f} ({p_over_1h*100:.2f}%)")
print(f"Probabilidad de NO ser abrumado en 24h: {p_no_over_day:.4f} ({p_no_over_day*100:.2f}%)")
print(f"Probabilidad de ser abrumado en algún momento del día: {p_over_day:.4f} ({p_over_day*100:.2f}%)")
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
# Datos
media_llegada = 10  # minutos
t = 5  # queremos que llegue en 5 min o menos

# Cálculo con distribución exponencial
prob = 1 - math.exp(-t / media_llegada)

print(f"Probabilidad de que el próximo cliente llegue en ≤5 minutos: {prob:.4f} ({prob*100:.2f}%)")
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
beta = 10  # media de tiempo entre clientes (minutos)
t = 15     # tiempo mínimo sin clientes para tomar descanso

# Probabilidad de poder tomar descanso (T >= 15)
prob_break = math.exp(-t / beta)

print(f"Probabilidad de que un empleado pueda tomar descanso: {prob_break:.4f} ({prob_break*100:.2f}%)")
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
media = 150  # gramos
desviacion = 10  # gramos

# Probabilidad de que el peso esté entre 140 y 160g
prob = norm.cdf(160, media, desviacion) - norm.cdf(140, media, desviacion)

print(f"Probabilidad de que el peso esté entre 140 y 160g: {prob:.4f} ({prob*100:.2f}%)")
```

**Challenge 6**

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
beta = 50  # media en horas
t = 30     # horas

# Probabilidad de fallo en las primeras 30h
prob_fail = 1 - math.exp(-t / beta)

print(f"Probabilidad de que falle en ≤30 horas: {prob_fail:.4f} ({prob_fail*100:.2f}%)")
```
